```python
import numpy as np
import pandas as pd
from scipy.linalg import solve
import random
```

```python
def Max(A):
    im=0
    jm=1
    amax=A[im,jm]
    for i in range(len(A)):
        for j in range(i+1,len(A)):
            if abs(A[i,j])>abs(amax):
                amax=A[i,j]
                im=i
                jm=j
    return im,jm
```

```python
def GC(A,eig):
    cir = np.array([[0.0 for j in range(2)] for i in range(len(A))])
    for i in range(len(A)):
        sum=0
        for j in range(len(A)):
            sum=sum+abs(A[i,j])
        cir[i,0]=A[i,j]
        cir[i,1]=sum-abs(A[i,i])
    for i in range(len(A)):
        if abs(cir[i,0]-eig)<=cir[i,1]:
            return 1
    return 0
```
